```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
df.head(5)
```

### Класс билетов

```python
fig = plt.figure(figsize=(18,12), dpi=1600) 
plt.subplot2grid((3,3),(0,0))
df['Pclass'].value_counts().plot(kind='barh', color='red', alpha=.5)
plt.title("Class Distribution")
plt.subplot2grid((3,3),(0,1))
df[df.Sex == 'male']['Pclass'].value_counts().plot(kind='barh', label='Male', alpha=.5)
df[df.Sex == 'female']['Pclass'].value_counts().plot(kind='barh', color='red', label='Female', alpha=.5)
plt.legend()
plt.subplot2grid((3,3),(0,2))
df[df.Survived == 1]['Pclass'].value_counts().plot(kind='barh', label='Survived', alpha=.5)
df[df.Survived == 0]['Pclass'].value_counts().plot(kind='barh', color='red', label='Not Survived', alpha=.5)
plt.legend()
ax1 = plt.subplot2grid((3,3),(1,0), colspan=2)
ax1.set_xlim(-10, 90)
df[df['Pclass'] == 1]['Age'].dropna().plot(kind='kde')
df[df['Pclass'] == 2]['Age'].dropna().plot(kind='kde')
df[df['Pclass'] == 3]['Age'].dropna().plot(kind='kde')
plt.xlabel("Age")
plt.title("Age Distribution with classes")
plt.legend(('1st Class', '2nd Class','3rd Class'),loc='best')
plt.subplot2grid((3,3),(1,2),)
```

```python
pd.crosstab(df.Pclass, df.Sex).div(pd.crosstab(df.Pclass, df.Sex).sum(1),axis=0).plot(kind='barh', stacked=True, color=['blue', 'red'])
```

```python
pd.crosstab(df.Pclass, df.Survived).plot(kind='barh', stacked=True, color=['blue', 'red'])
```
