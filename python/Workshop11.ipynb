```python
def derivative(f, l, r, h):
    x = numpy.linspace(l, r, h)
    return (f(x) - f(x - (r - l)/ h)) / ((r-l)/h)
```

```python
x = numpy.linspace(0, 2*pi, 1000)
d = derivative(sin, 0, 2*pi, 1000)
plot(x, d, 'b-o', x, np.cos(x), 'r-^')
```

```python
import pandas as pd
import datetime
```

```python
data = pd.read_csv('sem10/bikes.csv', index_col='datetime')
```

```python
data.index = pd.to_datetime(data.index)
data
```

```python
data[data.index >= datetime.datetime(2012, 12, 1, 0, 0, 0)]['count'].plot()
```

```python
data['temp'].resample('M', how=mean).plot()
```

```python
data['humidity'].resample('M', how=mean).plot()
```

```python
data['Hour'] = data.index.hour
data
```

```python
data[data.index < datetime.datetime(2012, 1, 1, 0, 0, 0)].groupby('Hour')['count'].mean().plot()
```
