```python
import urllib
from requests import get
from bs4 import BeautifulSoup
import string
import time
import threading
from Queue import Queue

urls = {}
    
def main(url, depth):
    try:
        print url
        req = get(url)
    except Exception as e:
        print str(e) + ' ERROR1 ' + url
    soup = BeautifulSoup(req.content.decode(req.encoding), "lxml")
    for raw_link in soup.findAll('a'):
            if 'href' in raw_link.attrs:
                link = raw_link['href']
                if link.startswith('/wiki/') and ':' not in link:
                    link = 'https://mo.wikipedia.org' + link
                    index = string.find(link, '#')
                    if index != -1:
                        link = link[:index]
                    index = string.find(link, '?')
                    if index != -1:
                        link = link[:index]
                    if link not in urls:
                        urls[link] = [depth + 1, raw_link.text, urllib.unquote(link)]
                        main(link, depth + 1)
                elif link.startswith('https://mo.wikipedia.org/wiki/') and ':' not in link.lstrip('https://'):
                    index = string.find(link, '#')
                    if index != -1:
                        link = link[:index]
                    index = string.find(link, '?')
                    if index != -1:
                        link = link[:index]
                    if link not in urls:
                        urls[link] = [depth + 1, raw_link.text, urllib.unquote(link)]
                        main(link, depth + 1)
```

```python
now = time.time()
if __name__ == '__main__':
    main('https://mo.wikipedia.org', 0)
print time.time() - now
```

```python
urls = {}
requests = Queue()
urls_q = Queue()
l = threading.Lock()
def parser(requests):
    while True:
        req = requests.get()
        soup = BeautifulSoup(req.content.decode(req.encoding), "lxml")
        for raw_link in soup.findAll('a'):
            if 'href' in raw_link.attrs:
                link = raw_link['href']
                if link.startswith('/wiki/') and ':' not in link:
                    link = 'https://mo.wikipedia.org' + link
                    index = string.find(link, '#')
                    if index != -1:
                        link = link[:index]
                    index = string.find(link, '?')
                    if index != -1:
                        link = link[:index]
                    l.acquire()
                    if link not in urls:
                        urls[link] = [raw_link.text, urllib.unquote(link)]
                        l.release()
                        urls_q.put(link)
                elif link.startswith('https://mo.wikipedia.org/wiki/') and ':' not in link.lstrip('https://'):
                    index = string.find(link, '#')
                    if index != -1:
                        link = link[:index]
                    index = string.find(link, '?')
                    if index != -1:
                        link = link[:index]
                    l.acquire()
                    if link not in urls:
                        urls[link] = [raw_link.text, urllib.unquote(link)]
                        l.release()
                        urls_q.put(link)
    
def zora(urls_q):
    while True:
        url = urls_q.get()
        try:
            print time.time(), url
            req = get(url)
            requests.put(req)
        except Exception as e:
            print str(e) + ' ERROR1 ' + url
```

```python
#tparser = threading.Thread(target=parser, args=(requests,))
#tzora = threading.Thread(target=zora, args=(urls_q,))
tparsers = [threading.Thread(target=parser, args=(requests,)) for i in range(2)]
tzoras = [threading.Thread(target=zora, args=(urls_q,)) for i in range(10)]
```

```python
for tzora in tzoras:
    tzora.start()
```
